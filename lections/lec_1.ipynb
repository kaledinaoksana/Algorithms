# A. Покраска деревьев

Вася и Маша участвуют в субботнике и красят стволы деревьев в белый цвет. Деревья растут вдоль улицы через равные промежутки в 1 метр. Одно из деревьев обозначено числом ноль, деревья по одну сторону занумерованы положительными числами 1,2 и т.д., а в другую — отрицательными -1, -2 и т.д.

Ведро с краской для Васи установили возле дерева P, а для Маши — возле дерева Q. Ведра с краской очень тяжелые и Вася с Машей не могут их переставить, поэтому они окунают кисть в ведро и уже с этой кистью идут красить дерево. Краска на кисти из ведра Васи засыхает, когда он удаляется от ведра более чем на V метров, а из ведра Маши — на M метров. 

Определите, сколько деревьев может быть покрашено.

```python
# P, V = map(int, input().split())
# Q, M = map(int, input().split())
```

```python
# функция правильного ответа
def answer_arr(P, V, Q, M):

    zero1: int = P
    delta1: int = V
    zero2: int = Q
    delta2: int = M

    if zero1 > zero2:
        zero1, zero2 = zero2, zero1
        delta1, delta2 = delta2, delta1
    
    min1 = zero1 - delta1
    max1 = zero1 + delta1
    
    min2 = zero2 - delta2
    max2 = zero2 + delta2

    if min1>min2:
        min1, max1, min2, max2 = min2, max2, min1, max1
        delta1, delta2 = delta2, delta1

    # print("1: [",min1,",", max1,"], \n2: [",min2,",", max2,"]")

    def res(min1, max1, min2, max2):
        
        if min1>min2:
            print("ERROR")
            return 0
    
        # одинаковое начало
        if min1==min2:
            return max(delta1,delta2)*2+1
        
        # интервал в интервале
        if max1>max2:
            return delta1*2+1

        # интервалы соприкасаются
        if max1==min2:
            return (delta1 + delta2 + 1) * 2 - 1
        
        #интервалы перекрываются
        if  max1>min2:

            if min2<0: 
                if max1>=0:
                    return (delta1 + delta2 + 1) * 2 - (abs(max1) + abs(min2) + 1) 
                else:
                    return (delta1 + delta2 + 1) * 2 - (abs(min2) - abs(max1) + 1)
            else: 
                return (delta1 + delta2 + 1) * 2 - (max1 - min2) - 1
        

        return (delta1 + delta2 + 1) * 2

    result = res(min1, max1, min2, max2)
    
    return result

def answer_set(P, V, Q, M):

    x1m = Q - M
    x2m = Q + M + 1
    x1v = P - V
    x2v = P + V + 1

    masha = set(range(x1m, x2m))
    vasua = set(range(x1v, x2v))

    return len(masha | vasua)
```

# C. Форматирование файла

Петя - начинающий программист. Сегодня он написал код из n строк.
К сожалению оказалось, что этот код трудно читать. Петя решил исправить это, добавив в различные места пробелы. А точнее, для i-й строки ему нужно добавить ровно a i пробелов.

Для добавления пробелов Петя выделяет строку и нажимает на одну из трёх клавиш: Space, Tab, и Backspace. При нажатии на Space в строку добавляется один пробел. При нажатии на Tab в строку добавляются четыре пробела. При нажатии на Backspace в строке удаляется один пробел.
Ему хочется узнать, какое наименьшее количество клавиш придётся нажать, чтобы добавить необходимое количество пробелов в каждую строку. Помогите ему!

### Формат ввода

Первая строка входных данных содержит одно целое положительное число – количество строк в файле.
Каждая из следующих n строк содержит одно целое неотрицательное число a – количество пробелов, которые нужно добавить в i-ю строку файла.

### Формат вывода

Выведите одно число – минимальное количество нажатий, чтобы добавить в каждой строке необходимое количество пробелов.

```python
def res_to_3(len):
    if len == 0:
        return 0
    if len == 1:
        return 1
    if len == 2:
        return 2
    if len == 3:
        return 2
    
def line(string):
    len = int(string % 4)
    main4 = int(string / 4)
    # print("len",len)
    # print("main",main4)
    return main4+res_to_3(len)


p = int(input())

res = 0
for i in range(p):
    str_res = line(int(input()))
    # print(str_res)
    res += str_res


print(res)
```

# D. Слоны и ладьи

На шахматной доске стоят слоны и ладьи, необходимо посчитать, сколько клеток не бьется ни одной из фигур.
Шахматная доска имеет размеры 8 на 8. Ладья бьет все клетки горизонтали и вертикали, проходящих через клетку, где она стоит, до первой встретившейся фигуры. Слон бьет все клетки обеих диагоналей, проходящих через клетку, где он стоит, до первой встретившейся фигуры.

### Формат ввода

В первых восьми строках ввода описывается шахматная доска. Первые восемь символов каждой из этих строк описывают состояние соответствующей горизонтали: символ B (заглавная латинская буква) означает, что в клетке стоит слон, символ R — ладья, символ * — что клетка пуста. После описания горизонтали в строке могут идти пробелы, однако длина каждой строки не превышает 250 символов. После описания доски в файле могут быть пустые строки.

### Формат вывода

Выведите количество пустых клеток, которые не бьются ни одной из фигур.

```python
reader = open('input.txt', 'r')
data = [reader.readline().strip()[:8] for _ in range(8)]
reader.close()


# Создаем пустой список для хранения координат символов 'R' и 'B'
coordinates = {"R":[],"B":[]}
board = [['*' for _ in range(8)] for _ in range(8)]

# Проходим по строкам и их символам для поиска символов 'R' и 'B'
for row_index, row in enumerate(data):
    for col_index, char in enumerate(row):
        if char == 'R':
            board[row_index][col_index] = "R"
            coordinates["R"].append((row_index, col_index))
        elif char == 'B':
            board[row_index][col_index] = "B"
            coordinates["B"].append((row_index, col_index))
            
coordinates
```

```python
def is_under_attack(row, col):
 
    # Проверка для ладьи по вертикали
    # Проверка вверх
    for i in range(row - 1, -1, -1):  
        if board[i][col] == 'B':  
            break
        if board[i][col] == 'R':  
            return True
    
     # Проверка вниз
    for i in range(row + 1, 8): 
        if board[i][col] == 'B':
            break
        if board[i][col] == 'R':
            return True

    # Проверка для ладьи по горизонтали
    # Проверка влево
    for j in range(col - 1, -1, -1):  
        if board[row][j] == 'B':
            break
        if board[row][j] == 'R':
            return True
    # Проверка вправо
    for j in range(col + 1, 8):  
        if board[row][j] == 'B':
            break
        if board[row][j] == 'R':
            return True

    # Проверка для слона
    # Первая диагональ
    for i, j in zip(range(row-1, -1, -1), range(col-1, -1, -1)):
        if board[i][j] == 'B':
            return True
        if board[i][j] != ('*' or '_'):
            break

    for i, j in zip(range(row+1, 8), range(col+1, 8)):
        if board[i][j] == 'B':
            return True
        if board[i][j] != ('*' or '_'):
            break

    # Вторая диагональ
    for i, j in zip(range(row-1, -1, -1), range(col+1, 8)):
        if board[i][j] == 'B':
            return True
        if board[i][j] != ('*' or '_'):
            break

    for i, j in zip(range(row+1, 8), range(col-1, -1, -1)):
        if board[i][j] == 'B':
            return True
        if board[i][j] != ('*' or '_'):
            break

    return False


count = 0

for row in range(8):
    for col in range(8):
        # Если клетка пуста и не бьется ни ладьей, ни слоном, увеличиваем счетчик
        if board[row][col] == '*' and not is_under_attack(row, col):
            board[row][col] = '_'
            count += 1

print(count)
```
